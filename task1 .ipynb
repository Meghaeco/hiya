**STEP-1 Importing Data Set**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("http://bit.ly/w-data")
```

** **STEP-1 Visualizing Data set**

```python
sns.jointplot(df["Hours"], df["Scores"], kind = "reg")
plt.show()
```

the relationship between scores and hours is linear. so we will use here linear regression model


**STEP-3 Data Preparation**

```python
df.head()
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, 1].values
```

```python
x
```

```python
y
```

**STEP-4 Training The Algorithm**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

regressor = LinearRegression()

regressor.fit(x_train.reshape(-1,1), y_train)
```

**STEP-5 Plotting The Regression Line**

```python
line = regressor.coef_*x+regressor.intercept_
plt.scatter(x, y)
plt.plot(x, line,color='black')
plt.show()
```

**STEP-6 Making Prediction**

```python
print(x_test)
y_pred = regressor.predict(x_test)
```

**STEP-7 Comparing Actual Data VS Predicted Model**

```python
x_test
```

```python
y_pred
```

```python
comp = pd.DataFrame({'Actual':[y_test], 'Predicted':[y_pred]})
comp
```

**WHAT WILL BE PREDICTED SCORE IF A STUDENT STUDIES FOR 9.25hrs/DAY**

```python
hours = 9.25
test = np.array([hours])
test = test.reshape(-1,1)
own_pred = regressor.predict(test)

print("No of hours studied per day:{}".format(hours))
print("predicted score:{}".format(own_pred[0]))
```
